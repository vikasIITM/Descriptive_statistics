**Function**

```python
from scipy.stats import skew,kurtosis

# For temp data
def func(yr1,yr2):
  df=res_df.loc[:,str(yr1):str(yr2)]
  df_min=df.min()
  df_max=df.max()
  df_mean=df.mean()
  df_var=df.var()
  df_std=df.std()
  df_skew=skew(df, axis=0, bias=True)
  df_kurtosis = kurtosis(df, axis=0, bias=True)
  Q1 = np.percentile(df, 25, interpolation = 'midpoint')
  Q3 = np.percentile(df, 75, interpolation = 'midpoint')
  # Interquaritle range (IQR)
  IQR = Q3 - Q1
  d = {'Min': df_min, 'Max': df_max,'Mean':df_mean,'Var':df_var,'Std':df_std,'Skewness':df_skew,'Kurtosis':df_kurtosis,'IQR':IQR}
  df = pd.DataFrame(data=d)
  return df

# Precipitation data
def func_p(yr1,yr2):
  df1=res_df1.loc[:,str(yr1):str(yr2)]
  df1_min=df1.min()
  df1_max=df1.max()
  df1_mean=df1.mean()
  df1_var=df1.var()
  df1_std=df1.std()
  df1_skew=skew(df1, axis=0, bias=True)
  df1_kurtosis = kurtosis(df1, axis=0, bias=True)
  Q1 = np.percentile(df1, 25, interpolation = 'midpoint')
  Q3 = np.percentile(df1, 75, interpolation = 'midpoint')
  # Interquaritle range (IQR)
  IQR1 = Q3 - Q1
  d = {'Min': df1_min, 'Max': df1_max,'Mean':df1_mean,'Var':df1_var,'Std':df1_std,'Skewness':df1_skew,'Kurtosis':df1_kurtosis,'IQR':IQR1}
  df1 = pd.DataFrame(data=d)
  return df1
```

```python
# For temp data
def func_l(Long,lat,yr1,yr2):
    location=res_df.loc[(res_df['Longitude'] == Long) & (res_df['latitude'] == lat)]
    df=location.loc[:,str(yr1):str(yr2)]
    x=np.arange(yr1,yr2+1)
    y=np.array(df).flatten()
    return plt.scatter(x,y)

# For precipitation data
def func_l_p(Long,lat,yr1,yr2):
    location=res_df1.loc[(res_df1['Longitude'] == Long) & (res_df1['latitude'] == lat)]
    df=location.loc[:,str(yr1):str(yr2)]
    x=np.arange(yr1,yr2+1)
    y=np.array(df).flatten()
    return plt.scatter(x,y)
```

```python
#For temp data
df_ll = df.drop(['latitude','Longitude'],axis = 1)
X = df_ll.mean(axis = 1)
ndf = df.iloc[:,:2]
X = pd.concat([ndf,X],axis = 1)
X.rename(columns = {0:'Mean'},inplace = True)

std= df_ll.std(axis=1)
std = pd.concat([ndf,std],axis = 1)
std.rename(columns = {0:'Std'},inplace = True)

min= df_ll.min(axis=1)
min= pd.concat([ndf,min],axis = 1)
min.rename(columns = {0:'Min'},inplace = True)

max= df_ll.max(axis=1)
max= pd.concat([ndf,max],axis = 1)
max.rename(columns = {0:'Max'},inplace = True)

var= df_ll.var(axis=1)
var= pd.concat([ndf,var],axis = 1)
var.rename(columns = {0:'Var'},inplace = True)


# For Precipitation data
df_llp = df1.drop(['latitude','Longitude'],axis = 1)
X1 = df_llp.mean(axis = 1)
ndf1 = df1.iloc[:,:2]
X1 = pd.concat([ndf1,X1],axis = 1)
X1.rename(columns = {0:'Mean'},inplace = True)

std1= df_llp.std(axis=1)
std1 = pd.concat([ndf1,std1],axis = 1)
std1.rename(columns = {0:'Std'},inplace = True)

min1= df_llp.min(axis=1)
min1= pd.concat([ndf1,min1],axis = 1)
min1.rename(columns = {0:'Min'},inplace = True)

max1= df_llp.max(axis=1)
max1= pd.concat([ndf1,max1],axis = 1)
max1.rename(columns = {0:'Max'},inplace = True)

var1= df_llp.var(axis=1)
var1= pd.concat([ndf1,var1],axis = 1)
var1.rename(columns = {0:'Var'},inplace = True)
```

```python
def Mean(data):
  fig,ax=plt.subplots(figsize=(5,5))
  plt.scatter(x=X['Longitude'],y = X['latitude'],c = X['Mean'])
  plt.xlabel('Longitude')
  plt.ylabel('Latitude')
  shapefile.plot(ax=ax,color='red')
  plt.colorbar()

def Meanp(data):
  fig,ax=plt.subplots(figsize=(5,5))
  plt.scatter(x=X1['Longitude'],y = X1['latitude'],c = X1['Mean'])
  plt.xlabel('Longitude')
  plt.ylabel('Latitude')
  shapefile.plot(ax=ax,color='red')
  plt.colorbar()

from sklearn.cluster import KMeans
def cluster(data):
     k_values = list(range(1, 11))
     sse_values = []
     for k in k_values:
       kmeans = KMeans(n_clusters=k)
       kmeans.fit(X)
       sse = kmeans.inertia_
       sse_values.append(sse)
     plt.plot(k_values, sse_values)
     plt.xlabel('k')
     plt.ylabel('SSE')
     plt.show() 
     kmeans = KMeans(n_clusters=4)
     kmeans.fit(X)
     y_pred = kmeans.predict(X)
     df['cluster'] = y_pred
     fig,ax=plt.subplots(figsize=(5,5))
     plt.scatter(x=df['Longitude'],y = df['latitude'],c = df['cluster'])
     plt.xlabel('Longitude')
     plt.ylabel('Latitude')
     shapefile.plot(ax=ax,color='red')
     plt.colorbar()

def clusterp(data):
     k_values = list(range(1, 11))
     sse_values = []
     for k in k_values:
       kmeans = KMeans(n_clusters=k)
       kmeans.fit(X1)
       sse = kmeans.inertia_
       sse_values.append(sse)
     plt.plot(k_values, sse_values)
     plt.xlabel('k')
     plt.ylabel('SSE')
     plt.show() 
     kmeans = KMeans(n_clusters=4)
     kmeans.fit(X1)
     y_pred = kmeans.predict(X1)
     df1['cluster'] = y_pred
     fig,ax=plt.subplots(figsize=(5,5))
     plt.scatter(x=df1['Longitude'],y = df1['latitude'],c = df1['cluster'])
     plt.xlabel('Longitude')
     plt.ylabel('Latitude')
     shapefile.plot(ax=ax,color='red')
     plt.colorbar()




def Std(data):
  fig,ax=plt.subplots(figsize=(5,5))
  plt.scatter(x=std['Longitude'],y = std['latitude'],c = std['Std'])
  plt.xlabel('Longitude')
  plt.ylabel('Latitude')
  shapefile.plot(ax=ax,color='red')
  plt.colorbar()


def Stdp(data):
  fig,ax=plt.subplots(figsize=(5,5))
  plt.scatter(x=std1['Longitude'],y = std1['latitude'],c = std1['Std'])
  plt.xlabel('Longitude')
  plt.ylabel('Latitude')
  shapefile.plot(ax=ax,color='red')
  plt.colorbar()

def Min(data):
  fig,ax=plt.subplots(figsize=(5,5))
  plt.scatter(x=min['Longitude'],y = min['latitude'],c = min['Min'])
  plt.xlabel('Longitude')
  plt.ylabel('Latitude')
  shapefile.plot(ax=ax,color='red')
  plt.colorbar()

def Minp(data):
  fig,ax=plt.subplots(figsize=(5,5))
  plt.scatter(x=min1['Longitude'],y = min1['latitude'],c = min1['Min'])
  plt.xlabel('Longitude')
  plt.ylabel('Latitude')
  shapefile.plot(ax=ax,color='red')
  plt.colorbar()
  
def Max(data):
  fig,ax=plt.subplots(figsize=(5,5))
  plt.scatter(x=max['Longitude'],y = max['latitude'],c = max['Max'])
  plt.xlabel('Longitude')
  plt.ylabel('Latitude')
  shapefile.plot(ax=ax,color='red')
  plt.colorbar()

def Maxp(data):
  fig,ax=plt.subplots(figsize=(5,5))
  plt.scatter(x=max1['Longitude'],y = max1['latitude'],c = max1['Max'])
  plt.xlabel('Longitude')
  plt.ylabel('Latitude')
  shapefile.plot(ax=ax,color='red')
  plt.colorbar()

def Var(data):
  fig,ax=plt.subplots(figsize=(5,5))
  plt.scatter(x=var['Longitude'],y = var['latitude'],c = var['Var'])
  plt.xlabel('Longitude')
  plt.ylabel('Latitude')
  shapefile.plot(ax=ax,color='red')
  plt.colorbar()

def Varp(data):
  fig,ax=plt.subplots(figsize=(5,5))
  plt.scatter(x=var1['Longitude'],y = var1['latitude'],c = var1['Var'])
  plt.xlabel('Longitude')
  plt.ylabel('Latitude')
  shapefile.plot(ax=ax,color='red')
  plt.colorbar()
```
